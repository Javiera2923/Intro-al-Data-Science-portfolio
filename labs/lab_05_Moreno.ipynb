1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
x  = df ['month_number'] #datos para el eje x
y = df['total_profit'] #datos para el eje y


#Gráfico lineal
plt.plot(
    x, # eje x
    y, # eje y 
    label='linea', # etiquetado que tiene el graf
    color="red", # color
    linewidth=2 # tamano de la curva
)

#plt.legend() # agregar etiquetado (puedo poner el label=linea pero si no pongo este comando no lo muestra)
plt.title("Beneficio total por cada mes") # agregar titulo
plt.xlabel('mes') # nombre eje x
plt.ylabel('beneficio total') # nombre eje y
plt.show() # mostrar grafico




# grafico de dispersión

plt.figure(figsize=(10, 6))

sns.scatterplot( #gráfico de dispersión - de ptos. 
    x  = df ['month_number'],
    y = df['total_profit'],
   
    data=df,
    #palette=palette
)   
plt.title("Beneficio total por cada mes") # agregar titulo
plt.xlabel('mes') # nombre eje x
plt.ylabel('beneficio total') # nombre eje y
plt.show() # mostrar grafico

```

2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
#Entiendo que hay que graficar la venta de cada producto en los 12 meses 
#Cada producto será una línea distinta.

#datos para el eje x
x  = df ['month_number']

#datos para el eje y de cada gráfico
facecream_sales = df ['facecream']
facewash_sales = df ['facewash']
toothpaste_sales = df ['toothpaste']
bathingsoap_sales = df ['bathingsoap']
shampoo_sales = df ['shampoo']
moisturizer_sales = df ['moisturizer']

#Gráfico facecream
plt.plot(
    x, # eje x
    facecream_sales , # eje y 
    label='facecream', # etiquetado que tiene el graf
    color="red", # color
    linewidth=2 # tamano de la curva
)

#Gráfico facewash
plt.plot(
    x, # eje x
    facewash_sales , # eje y 
    label='facewash', # etiquetado que tiene el graf
    color="blue", # color
    linewidth=2 # tamano de la curva
)

#Gráfico toothpaste
plt.plot(
    x, # eje x
    toothpaste_sales , # eje y 
    label='toothpaste', # etiquetado que tiene el graf
    color="green", # color
    linewidth=2 # tamano de la curva
)

#Gráfico bathingsoap
plt.plot(
    x, # eje x
    bathingsoap_sales , # eje y 
    label='bathingsoap', # etiquetado que tiene el graf
    color="yellow", # color
    linewidth=2 # tamano de la curva
)

#Gráfico shampoo
plt.plot(
    x, # eje x
    shampoo_sales , # eje y 
    label='shampoo', # etiquetado que tiene el graf
    color="brown", # color
    linewidth=2 # tamano de la curva
   
)

#Gráfico moisturizer
plt.plot(
    x, # eje x
    moisturizer_sales , # eje y 
    label='moisturizer', # etiquetado que tiene el graf
    color="black", # color
    linewidth=2 # tamano de la curva
)


plt.legend() # agregar etiquetado
plt.title("Ganancia mes a mes de cada producto") # agregar titulo
plt.xlabel('Mes') # nombre eje x
plt.ylabel('Producto') # nombre eje y

plt.show() # mostrar grafico
```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python

#datos para el eje y 
facecream_sales = df ['facecream']
facewash_sales = df ['facewash']

plt.subplot(1, 1, 1) # sub-ventana
plt.bar([a-0.25 for a in x], facecream_sales, width= 0.25, color='blue', label = 'Facecream sales', align='edge') #barras para los datos de facecream (en azul)
plt.bar([a+0.25 for a in x], facewash_sales, width= -0.25, color='red',label = 'Facewash sales', align='edge') #barras para los datos de facewash (en rojo)
plt.ylabel("Número de ventas por producto") #leyenda para eje y
plt.xlabel('Mes') #leyenda eje x
plt.legend(loc='upper left') #leyenda arriba a la izquierda

plt.xticks(x) #etiquetas en el eje x
plt.title('Datos de venta de productos de "Face Cream" y "Face Wash"')
plt.show()
```

4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
df = df.drop([ 'month_number', 'total_units', 'total_profit'], axis=1) #Le quito estas 3 columnas que anuncio aquí



plt.figure(figsize=(10, 6))
sns.boxplot(data=df) #instancio boxplot y le doy el conjunto de datos (un boxplot para cada columna)
plt.show()
```

5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
#Etiquetas de mi gráfico:
labels = ['Face cream', 'Fase wash', 'Tooth paste', 'Bathing soap', 'Shampoo', 'Moisturizer']

#suma de ventas totales del año pasado de cada producto:
sales = [df ['facecream'].sum(), 
         df ['facewash'].sum(), 
         df ['toothpaste'].sum(), 
         df ['bathingsoap'].sum(), 
         df ['shampoo'].sum(), 
         df ['moisturizer'].sum()]

plt.axis("equal")
plt.pie(sales, labels=labels, autopct='%1.1f%%') #autopct muestra el valor porcentual de las ventas totales de cada producto
plt.legend(loc='upper left') #posición leyenda
plt.title('Ventas totales del año pasado por cada producto') #título
plt.show()
```
